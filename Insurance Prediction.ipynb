# Insurance Prediction with Linear Regression

# Import Necessary Files For Linear Regression Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Insurance_Dataset = pd.read_csv("Insurance Dataset - Simple Linear Regression.csv")
Insurance_Dataset
```

# Create a LinearRegression model

```python
reg = LinearRegression()
x = Insurance_Dataset[['Age']]
y = Insurance_Dataset['Premium']

# Fit the linear regression model to the data
reg.fit(x, y)

# Now, the model is trained and ready for predictions
```

```python
print('The COEF:',reg.coef_)
print('The INTERCEPT:',reg.intercept_)
```

# Prediction

```python
user_age = float(input("Enter age: "))

# Predict premium based on user input
predicted_premium = reg.predict(np.array([[user_age]]))

# Display the predicted premium
print(f"Predicted Premium for {user_age} Years of Age: {predicted_premium[0]}")

# Visualize the Results (optional)
plt.scatter(x, y, label='Actual Data')
plt.plot(x, reg.predict(x), color='red', linewidth=3, label='Linear Regression')
plt.scatter(user_age, predicted_premium, color='green', label='Predicted Value', marker='*', s=100)
plt.title("Linear Regression: Age vs. Predicted Premium")
plt.xlabel("Age")
plt.ylabel("Predicted Premium")
plt.legend()
plt.show()
```
